```python
def sci_nearest_power10(x):
    """
    Return string like '1.0 × 10^3'
    where exponent is nearest power of 10.
    """
    if x == 0:
        return r'$0$'

    exp = int(np.round(np.log10(abs(x))))
    mantissa = x / 10**exp

    return rf'${mantissa:.1f}\times10^{{{exp}}}$'
```

```python
def func_mi(m, r, k2 = 0.07):
    
    return k2 * m * r * r

def func_Jdot_mb(m, r, omega, const = 0.5e-28, alpha = 4.0, beta = 3.0, k2 = 0.076, f2 = 1.0):
    
    return -const * k2 * f2 * m * (rsun**4) * ((r/rsun)**alpha) * (omega**beta) * yr_to_sec

def func_Jdot_acc(m, mdot, r):
    
    return mdot * ((G*m/r)**(0.5)) * r

def func_omega_crit(m, r):
    
    return (G*m/(r**3))**(0.5)
```

```python
def roche_lobe_radius(q, a_Rsun):
        
    q23 = q**(2/3)
    q13 = q**(1/3)
    
    rl_over_a = 0.49 * q23 / (0.6 * q23 + np.log(1 + q13))
    
    return rl_over_a * a_Rsun


def outer_lobe_radius(q, r_l1_Rsun):

    fact1 = 0.441*(q**(0.325))
    fact2 = 1.0 + 0.412*(q**(0.8))

    fact = 1.0 + fact1/fact2

    return r_l1_Rsun * fact
```
